Q.2 Explain the diff. between mutable and immutable data types with example?

->> A mutable data type is one whose state or contents can be modified after it is created. Examples - Lists, Dictionaries and sets. Whereas, An immutable data type is one whose state or contents cannot be changed after it is created. Once an object of an immutable type is created, it cannot be altered directly. Examples - Strings, tuples and integers.

Q.3 What are the main differences b/w lists and tuples in python?

->> Lists and Tuples are both used to store collections of items, but they have several important differences.

Lists

1.Lists are mutable, meaning their contents can be changed.

2.Lists are created using square brackets[].

3.Lists have a large number of built-in methods, such as append(), extend(), remove(), pop(), and sort(), which allow modification of the list.

4.Lists are typically used when the collection of items needs to be modified frequently, or when the collection is dynamic. Ex- elements are added or removed.

5.Lists are not hashable because they are mutable, so they cannot be used as keys in dictionaries.

Tuples

1.Tuples are immutable, meaning once they are created.

2.Tuples are created using parentheses().

3.Tuples have fewer built-in methods, only count() and index() are available.

4.Tuples are used when the collection of items should not change. Ex- for use in dictionaries or sets.

5.Tuples are hashable (as long as all their elements are also hashable).

Q.4 Describe how dictionaries store data?

->> Dictionaries in Python are a type of associative array or hash map that store data in key-value pairs. The key is used to look up the associated value, and each key must be unique within the dictionary. A dictionary is composed of keys and their corresponding values. Under the hood, Python dictionaries use a hash table to store data. Dictionaries in Python are unordered collections (prior to Python 3.7). In earlier versions, the order of items in a dictionary could not be relied upon.

Q. 5 Why might you use a set instead of list in Python?

->> A set automatically ensures that all elements are unique. If you try to add a duplicate element, it will not be added. A list allows duplicate elements, which means you need to manually handle duplicates if you don't want them.

Q. 6 What is a string in Python, and how is it different from a list?

->> a string is a sequence of characters enclosed within quotes (either single (') or double (") quotes). Strings are used to represent textual data, such as words, sentences, or even paragraphs. They are immutable objects, meaning once created, their contents cannot be changed. It is different from list in mutability, methods, use cases and mixing.

Q. 7 What is a hash table, and how does it relate to dictionaries in Python?

->>  Hash table is a data structure that allows for efficient storage and retrieval of data using a hash function. It stores data in key-value pairs, where each key is mapped to a value.

Q. 9 Explain why strings are immutable in Python?

->> Strings in Python are immutable, meaning once a string object is created, it cannot be changed or modified. This immutability is a key characteristic of strings in Python and serves several purposes in the language design, with performance, safety, and consistency being some of the most important reasons. Immutable strings can be safely shared across multiple threads without risk of modification. Since strings can’t be changed, you don’t have to worry about unexpected side effects.

Q. 10 Describe a scenario where using a tuple would be preferable over a list?

->> In Python, a tuple would be preferable over a list in scenarios where immutability (the ability to ensure that the data cannot be changed) is important. Tuples are more efficient than lists in terms of memory and performance when the collection of items is fixed and should not be modified.

Q. 11 How do sets handle duplicate values in Python?

->> In Python, sets automatically remove duplicates. When you try to add a duplicate value to a set, it simply ignores the new value and does not add it. A set only contains unique elements, so any repeated elements are automatically discarded.

Q. 12 How does the "in" keyword work differently for lists and dictionaries?

->> The in keyword is used in both lists and dictionaries in Python to check for membership, but it works differently depending on the data structure. When used with a list, the in keyword checks if a specific element exists in the list. When used with a dictionary, the in keyword checks if a key exists in the dictionary, not the value. It performs a hash-based lookup.

Q. 13 Can you modify the elements of a tuple? Explain why or why not.

->> No, you cannot modify the elements of a tuple in Python because tuples are immutable. Once a tuple is created, its contents cannot be changed, added to, or removed. If you attempt to modify an element of a tuple, Python will raise a TypeError ('tuple' object does not support item assignment).

Q. 14 In what situations are lists preferred over dictionaries?

->> In Python, lists and dictionaries are both used to store collections of data, but they serve different purposes based on their characteristics. Lists are preferred over dictionaries in situations where order matters, When You Need to Store Multiple Items of Any Type, You Need to Use Indexes to Access Elements, When Data Modifications (Appends, Inserts, etc.) are Needed Frequently etc.

Q. 15 Explain the difference b/w a list and a dictionary in terms of data retrieval.

->> Lists in Python are ordered collections of elements that are accessed by their index (a position-based identifier, starting at 0), whereas Dictionaries in Python are unordered collections of key-value pairs. To retrieve a value from a dictionary, you use the key associated with that value. To access an element in a list, you need to know the index of the element you want to retrieve and on the other hand, ou access the value directly by its key, which can be any immutable object (e.g., strings, numbers, tuples).
